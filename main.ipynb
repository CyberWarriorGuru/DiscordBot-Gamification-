```python
with open('new_dataset.yml', 'r') as file:
    data = yaml.safe_load(file)
```

```python
convos = data['conversations']
```

```python
output = []
```

```python
for convo in convos:
    completion = ''
    for i, dialog in enumerate(convo):
        if i == 0:
            prompt = dialog
            prompt = prompt.replace("\xa0", " ")
        else:
            completion += " " + dialog
            completion = completion.replace("\xa0", " ")
    completion = completion.strip()
    line = {'prompt': prompt, 'completion': completion}    
    output.append(line)
print(output)
```

```python
with open('new_dataset-d.jsonl', 'w') as outfile:
    for i in output:
        json.dump(i, outfile)
        outfile.write('\n')
```
